#### 检查系统中是否已安装旧版mysql

    rpm -qa | grep -i mysql

    卸载旧版：
    
    rpm -ev <mysql版本>
    
    倘若提示有依赖关系，则可以使用yum进行卸载
    
    yum -y remove <mysql版本>

#### 下载mysql
    
    选择通用版：linux generic
            
            Linux - Generic (glibc 2.12) (x86, 64-bit), Compressed TAR Archive
            
    （将链接复制到迅雷能加快下载速度）
    下载完成后通过Xftp6工具传输到linux中
    
    
#### 解压
    
     tar -xvf mysql-8.0.21-linux-glibc2.12-x86_64.tar.xz 
    

#### 重命名解压后的文件夹
       
    重命名为：mysql
    
    mv mysql-8.0.21-linux-glibc2.12-x86_64.tar.xz mysql
    
    
#### 移动mysql至/usr/local/目录

    mv mysql /usr/local
    

#### 创建mysql用户及用户组
    
    新建一个用户mysql，以及用户组mysql，由该用户对mysql数据库进行专门管理。(mysql用户及用户组有可能已被数据库自动创建。)
    
    创建用户组：
        groupadd mysql
        
    创建用户，并将该用户添加到mysql用户组
        useradd mysql -g mysql
        
     设置mysql用户的登录密码：
         passwd <密码>
         
         
#### 创建data存储文件夹

     新建/usr/local/mysql/data 文件夹（进入/usr/local/mysql目录）
         mkdir data
     

#### 初始化数据库（会生成随机密码，初次登录时使用）

    /usr/local/mysql/bin/mysqld --initialize --user=mysql --basedir=/usradir=/usr/local/mysql/data/



#### 设置数据库归属

     修改mysql数据库的归属：
     
         mysql数据库文件所有者设为root用户，data文件夹所属用户设为mysql,以及mysql组用户
         
         chown -R root:root mysql
         chown -R mysql:mysql data 
     
     
#### 编辑/etc/my.conf文件

    新建或编辑/etc/my.cnf文件
    
    vim /etc/my.cnf
    
    编辑内容如下：
    
    [mysqld]
    basedir = /usr/local/mysql
    datadir = /usr/local/mysql/data
    port = 3306
    socket = /tmp/mysql.sock
    sql_mode = NO_ENGINE_SUBSTITUTION,STRICT_TRANS_TABLES
    
    
    修改my.cnf文件的权限：
    
    chmod 777 /etc/my.cnf
    
#### 编辑/etc/ld.so.conf文件
    
    指定mysql第三方库的路径
    
    vim /etc/ld.so.conf
    
    在末尾添加一行
    /usr/local/mysql/lib


 
#### 配置环境变量/etc/profile
    
    末尾添加一行：
    export PATH=$PATH:/usr/local/mysql/bin:/usr/local/mysql/lib 
    
    $使设置立即生效
     
    source /etc/profile 
    
   
#### 设置开机自启动

    
     cp /usr/local/mysql/support-files/mysql.server /etc/init.d/mysql
     
     chmod +x /etc/init.d/mysql 

     注册并查看各运行级别下mysql的运行设置：
     
     chkconfig --add mysql
     
     chkconfig --list mysql
     
     
     
#### 启动mysql服务
    
    service mysql start
    

#### 登录mysql
    
    mysql -u root -h localhost -p
    
    修改密码：
    
    alter user 'root'@'localhost' identified by '123456';
    
    退出：
    
    exit
