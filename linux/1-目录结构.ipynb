### 各目录功能

#### /bin
    binary缩写，用于存放最经常使用的命令
    
#### /sbin
    superuser bin,存放管理员用于管理系统的管理程序
    
#### /home
    存放普通用户的目录，每个用户对应一个文件夹
    
#### /root
    系统管理员的用户存放目录
    
#### /lib
    系统开机时需要用到的动态连接共享库
    
#### /lost+found
    系统意外关机时会生成一些文件
    
#### /etc
    系统管理需要的配置文件和子目录, 如mysql的my.config
    
#### /usr
    用户的应用程序和文件的存储位置
    
#### /boot
    启动linux时需要使用的文件
    
#### /proc
    这是一个虚拟目录，为系统内存的映射，用于获取系统信息
    
#### /srv
    service的缩写，存放着服务启动之后需要提取的数据
    
#### /sys
    系统内核文件
    
#### /tmp
    临时文件
    
#### /dev
    类似设备管理器，以文件的形式保存设备
    
#### /media
    U盘，光驱等设备挂载目录
    
#### /mnt
    用于挂载其他文件系统
    
#### /opt
    系统安装额外软件的存放目录
    
#### /usr/local
    另一个用于系统安装额外软件的存放目录，一般为源码编译安装的软件
    
#### /var
    存放不断扩充的文件，如日志文件等
    
#### /selinux
    安全子系统，可限制程序能访问的文件
