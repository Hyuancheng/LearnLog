### 系统运行级别配置文件
    /etc/inittab


### 七个运行级别
    
    0：关机
    1：单用户模式
    2：多用户状态没有网络服务
    3：多用户状态有网络服务
    4：保留级别
    5：图形界面
    6：系统重启


### 指定系统默认运行级别
    root直接在终端输入：
    init 级别号码(0123456)
    
    或者修改/etc/inittab 文件中最后一行的默认运行级别。
    
    如：
    
    id:5:initdefault:
    
    5表示当前默认运行级别为5

## 帮助指令

### man 获得帮助信息
    man [命令或配置文件]
    
    

### help 获取帮助信息




### 获取当前目录绝对路径
    pwd
    
    
    
### 显示当前目录文件的所有信息
    ls [可选项] 
    
    可选项为：
    -l 格式化列出
    -a 把隐藏文件也显示出来  
    
    
    
### 创建目录

#### 创建单级目录
    mkdir 绝对（/相对）路径
    
#### 创建多级目录
    mkdir -p 绝对（/相对）路径   
    
 
### 删除目录

#### 删除空文件夹
    rmdir 绝对（/相对）路径
    例如：rmdir /home/xid  
    仅当xid文件夹为空时，xid文件夹才会被删除
       
#### 删除非空目录
    rm -rf 绝对路径
    
    
### 创建空文件    
    创建一个名字为name的txt文件：
    touch name.txt
    
    也可以同时创建多个文件，文件名之间用空格隔开


### 拷贝文件
    cp 【可选项】 文件路径（相对/绝对）  目标路径
    
#### 拷贝单个文件
    如：拷贝/home/name.txt到 /home/next/目录下
    cp /home/name.txt  /home/next/
    
#### 拷贝整个文件夹（递归拷贝文件）
    带上-r参数
    如：拷贝整个/home/pws文件夹到/home/next/下
    cp -r /home/pws /home/next/
    
    当目标目录下已有相同名称文件夹，系统会对每个文件都会询问是否覆盖。
    倘若不想一一处理，想全部覆盖，则可以在cp前面加“/”
    /cp -r /home/pws /home/next/
    

### 删除文件

    rm 文件路径
    
    会提示是否确定删除
    
#### 强制删除

    rm -f 文件路径
    
    没有提示

### 移动或重命名文件
    mv 文件名 目标路径[文件名]
    
    当目标路径与移动文件处于同一目录时，mv指令即可以修改文件名称
    

### 以只读方式查看文件内容
    cat [选项] 路径/文件名
    
    可选项为：
    -n   显示行号
    
    倘若想分页查看，则可配合more使用
    cat -n 路径 | more
    
    按空格进行翻页

### 以全屏方式按页展示文件内容--more
    more 文件路径
    
    打开文件后的指令
    空格：下一页
    回车：下一行
    ctrl+B: 上一页
    ctrl+f: 下一页
    q: 退出more
    
    
    
  
### 按页查看文件内容--less
    less 文件路径
    逐步加载文件，在处理大型文件时具有很大优势
    
    文件内指令：
    空格：下一页
    回车：上一页
    q：退出
    
    less具备字符串查找功能，需要时可百度查看指令
    
    
    
### 输出重定向
    > 
    将列表内容覆盖到 文件名 中：
        ls -l > 文件名
    将文件1的内容覆盖到文件2中：
        cat 文件1 > 文件2 
        
   
   
### 追加内容
    >>
    将列表内容追加到 文件名 中：
        ls -l >> 文件名
    将文件1的内容追加到文件2中:
        cat 文件1 >> 文件2
    将字符串追加到 文件名 中：
        echo "字符串" >> 文件名
        
        
        
### 输出内容到控制台
    echo 内容
    
#### 输出当前环境变量的路径
    echo $PATH($末尾无需该符号) 

#### 输出内容到终端
    echo "字符串"


### 查看文件开头部分
    head 路径
    默认显示前10行
    
    可指定显示行数
    head -n [num] 路径
    
### 查看文件结尾部分
    tail 路径
    默认显示后10行
    
    指定显示行数
    tail -n [num] 路径
    
    实时监控文件变化
    tail -f 路径
    
    
    
        
        

### 查询历史指令
    history
    这会将所有历史指令罗列出来
    
    可以指定展示最近使用指令的数量
    history [num]
    
    执行展示出来的指令
    ![num]
    展示的指令有该指令的编号，只需是使用感叹号+该指令编号便可执行
