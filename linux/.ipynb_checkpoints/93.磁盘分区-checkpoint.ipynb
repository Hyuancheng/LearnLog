### 查看磁盘分区
    lsblk -f
    
    如果是在root用户下，该命令会列出四列信息，分别为：分区情况、分区类型、分区id、挂载点。
    
    
### 挂载硬盘 -- 以虚拟机为例
    1.虚拟机增加硬盘
     虚拟机的设置选项，添加硬盘scis，创建虚拟磁盘，选定磁盘大小，选择拆分成多个文件。
     
     重启虚拟机，输入查看分区指令，即可发现sdb这块新增的硬盘，但此时该硬盘并没有分区信息。
     
    2.对硬盘进行分区
     输入fdisk /dev/sdb 指令
     输入m 获取帮助信息
     输入n 增加一个新的分区
     输入p 设置成一个主分区
     输入1 编号为1
     输入w 将分区写入硬盘后退出
     此时，执行查看分区指令，发现硬盘sdb下面多了sdb1分区
     至此，分区创建完毕
     
    3.格式化分区
     mkfs -t ext4 /dev/sdb1
     此时，分区拥有id
     
    4.挂载
     在想挂载的地方新建一个文件夹，如/home/xiaoming/yingpan
     mount /dev/sdb1 /home/xiaoming/yingpan
     此时yingpan文件夹即为sdb1分区
     
    5.设置自动挂载
     当系统重启时，上面的挂载关系需要再次手动执行，若希望自动执行，则可修改fstab文件,该文件记录了分区及挂载点的情况
      vim /etc/fstab
      
      新增一行：
      /dev/sdb1  /home/xiaoming/yingpan      ext4   defaults  0 0
      表示将sdb1分区挂载到yingpan文件夹，类型为ext4 
      
      保存文件后，执行：
      mount -a 
      立即生效
      
    6.卸载分区
        umount 分区名称（/etc/sdb1）或者挂载目录（/home/xiaoming/yingpan）

### 磁盘情况查询
    df -lh
    
    
#### 查询指定目录的磁盘占用情况
    du [选项] 文件目录
    选项如下：
    -s: 指定目录占用大小汇总
    -h: 带计量单位
    -a: 含文件
    -max-depth = 1 子目录深度
    -c: 列出明细以及汇总值
    
    
#### 统计某个目录下含多少个文件（不含文件夹）
    ls -l 目录 | grep "^-" | wc -l
    
    "^-": 表示以"-"开头，因为以d开头的是文件夹而不是文件
    
#### 统计某个目录下含多少个文件夹
    将上述“-”改为“d”就行

#### 递归统计文件夹/文件数量
    带上R参数即可
    文件夹为“d”, 文件为“-”
    ls -lR 目录 | grep "^-" | wc -l


#### 以树的形式显示目录结构
    tree
    显示当前目录往下的结构
    
    tree 指定目录
