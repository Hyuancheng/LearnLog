### nginx是什么
    nginx是高性能的http和反向代理服务器。
    
    
    
### 反向代理

#### 正向代理
    
    客户端不直接访问目标服务器，而是将访问请求发送给中间的一个服务器（该服务器就是客户端的代理服务器），由该服务器向目标服务器发起请求。
    中间服务器提供的这种服务就叫正向代理。
    
    所谓“翻墙”，即通过代理服务器访问外网。

#### 反向代理

    在正向代理中，客户端需要配置代理服务器，即客户端清楚自己是通过另一台服务器访问目标服务器，是有感知的。
    在反向代理中，客户端直接访问“目标服务器”，但是我们需要的资源实际上并不在该服务器上，该服务器也只是一个代理服务器，它会将请求转发给后方资源服务器。在这种情况下，客户端对代理服务器的存在是无感知的，而这种代理方式也被称为反向代理。
    不难发现，正向代理与反向代理可以同时存在。
    反向代理服务器可以连接多台资源服务器，这样就可以大大强化资源服务器处理请求的能力。



### 负载均衡
    
    反向代理服务器可以连接多台资源服务器。当面临大量请求时，代理服务器可以根据各个资源服务器的运行情况将请求分发到不同服务器上，以减轻单台服务器的压力，这种处理方式即为负载均衡。




### 动静分离

    即将静态资源和动态资源分别存储到不同的资源服务器中，以达到效率最大化。
